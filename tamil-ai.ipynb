```python
from transformers import AutoTokenizer, AutoModelWithLMHead, pipeline
model_name = 'abinayam/gpt-2-tamil'
model = AutoModelWithLMHead.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
input_text = "ஒரு ஊரிலே ஒரு காக்கைக்கு"
max_len = 300
no_seq = 5
generator = pipeline('text-generation', model=model, tokenizer=tokenizer)
sequence = generator(input_text, max_length=max_len, num_return_sequences=no_seq)
print(sequence)
```

```python
# Load model directly
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

checkpoint = "Mr-Vicky-01/English-Tamil-Translator"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSeq2SeqLM.from_pretrained(checkpoint)


```

```python
def language_translator(text):
    tokenized = tokenizer([text], return_tensors='pt')
    out = model.generate(**tokenized, max_length=128)
    return tokenizer.decode(out[0],skip_special_tokens=True)

text_to_translate = "hardwork never fail"
output = language_translator(text_to_translate)
print(output)
```

```python
text_to_translate = "Artificial Intelligence is Revolutioning the World"
output = language_translator(text_to_translate)
print(output)
```
