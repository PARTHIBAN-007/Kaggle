<div style="background-color:#ffc3a0; color: #dd1818; text-align:center; font-size:25px; padding:15px; border-radius:25px; ">Data Analysis</div>

```python
i = 0
plt.figure(figsize=(12,10))
def hist_plot(col):
    global i
    i+=1
    plt.subplot(3,2,i)
    sns.histplot(x=df[col],kde=True)
    plt.xlabel(col)
    plt.ylabel(" ")
    plt.title(f"Distribution of {col}")



for columns in ['Age','Work Pressure','Job Satisfaction','Work Hours','Financial Stress']:
    hist_plot(columns)

plt.tight_layout()

plt.show()
```

```python
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'salmon'] ,startangle=0,explode=[0.05, 0.05])
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```
